```python
from datetime import datetime

def to_minutes(x):
    h,m = x.split(':')
    return int(h)*60 + int(m)

def findDayOfWeek(dateStr):
    date_object = datetime.strptime(dateStr, '%m/%d/%Y')
    dayOfWeek = date_object.strftime('%A')
    return dayOfWeek 

# findDayOfWeek('01/01/2006')

def flightNoStr(x):
    return str(int(x))

def classifyDelay(delay):
    if delay < -5:
        return 0 #flight is early
    elif delay > 5:
        return 1 #flight is delayed
    else:
        return 2 #flight is ontime
# def calculateDepTime()
```

```python
# Airport on time departure percentage data
# Replace 'your_file.xlsx' with the path to your Excel file
file_path = 'data/airline-on-time-arrival-pct/airline-on-time-pct-2023.xlsx'

# Load the Excel workbook
workbook = load_workbook(file_path)

# Get the names of all worksheets
sheet_names = workbook.sheetnames

# Read data from each worksheet
for sheet_name in sheet_names:
    # Access each worksheet by its name
    worksheet = workbook[sheet_name]
    print(f"Reading data from '{sheet_name}'")

    names = sheet_name.split()
    if len(names) > 1:
        if 'Marketing' in names[0]:
            print('Skipping this sheet, as it contains marketing on time arrival percentage')
            continue
        else:
            year = names[1]
    else:
        year = names[0]

    if int(year) < 2006:
        print('Skipping years earlier than 2006')
        continue

    # Iterate through all rows and columns
    it = 0
    airline_data = {}
    metadata_str = ['Reporting Operating Airline On-Time Arrival','On-Time Arrival Pct',
                    'Source: Bureau of Transportation Statistics', 'All Carriers',
                   'branded code-share partners', 'stopped codesharing', ' ceased operations',
                    'combined reporting', 'ranked based on', 'stopped reporting']
    for row in worksheet.iter_rows(values_only=True):
        row_str = " ".join(str(element) for element in row)
        
        skiprow = False
        for s in metadata_str:
            s = s.lower()
            if s in row_str.lower():
                skiprow= True
            if all(element == None for element in row):
                skiprow=True
        if skiprow:
            continue
        print(row_str)
        
        row_arr = list(row)
        
        airline_name = row_arr[1]
        
        if '*' in airline_name:
            airline_name = ''.join([char for char in airline_name if char != '*'])
            print('after removing start airline name {}'.format(airline_name))
        
        airline_name_first = airline_name.split()[0]

        substring_found = False
        
        for element in list_of_airlines:
            if airline_name_first.lower() == element.lower():
                airline_name_first = element
                substring_found = True
        
        # any(airline_name_first.lower() in element.lower() for element in list_of_airlines)
        
        if not substring_found:
            print('{} not in our dataset'.format(airline_name))
            continue
        
        airline_on_time_arrival_pct = round(float(row_arr[2]), 2)
        airline_data[airline_name_first] = airline_on_time_arrival_pct
    
    airline_on_time_arr_pct_data[year] = airline_data

# Close the workbook
workbook.close()
print(airline_on_time_arr_pct_data)
# json.dump(airport_on_time_dep_pct, indent=4)
```

```python
# adding minimum reported percentage for the years airlines did not report on time arrival percentage
minimum_on_time_arr_pct = {}
for year, airline_data in airline_on_time_arr_pct_data.items():
    # print(year)
    # print(airline_data)
    for airline in airline_data:
        if airline not in minimum_on_time_arr_pct:
            minimum_on_time_arr_pct[airline] = airline_data[airline]
        else:
            min = minimum_on_time_arr_pct[airline]
            current_val = airline_data[airline]
            if current_val < min:
                # min = current_val
                minimum_on_time_arr_pct[airline] = current_val

print('#### minimum percentage of on-time flights #####')
# print(minimum_on_time_arr_pct)

for year in airline_on_time_arr_pct_data:
    airline_data = airline_on_time_arr_pct_data[year].copy()
    # print(airline_data)
    for airline in list_of_airlines:
        if airline not in airline_data:
            airline_data[airline] = minimum_on_time_arr_pct[airline]
    airline_on_time_arr_pct_data[year] = airline_data

data_for_2024 = airline_on_time_arr_pct_data['2023'].copy()
airline_on_time_arr_pct_data['2024'] = data_for_2024      
print('### FINAL ON TIME PERCENTAGE DATA ####')
# print(airline_on_time_arr_pct_data)
```

```python
#put weather data, airport on time dep pct, because you're processing date and airport origin here. 
def addDepPercentage(dateStr, origin):
    year = str(datetime.strptime(dateStr, '%m/%d/%Y').year)
    airport_on_time_dep_rate = airport_ranking_data[year][origin]
    return airport_on_time_dep_rate

def addArrPercentage(dateStr, carrier_code):
    year = str(datetime.strptime(dateStr, '%m/%d/%Y').year)
    airport_on_time_dep_rate = airline_ranking_data[year][carrier_code]
    return airport_on_time_dep_rate
```

```python

# combinned_df = pd.DataFrame()
# if RawDataProcessing:
dataDir = 'data/arrivals/'
fileNames = os.listdir(dataDir)
dfList = []
for file in fileNames:
    filePath = dataDir + file
    print('Processing file {}'.format(filePath))
    if Path(filePath).is_dir():
        print('Skipping the directory {}'.format(filePath))
        continue
    raw_df = pd.read_csv(filePath)
    # raw_df.head()

    airline_df = pd.DataFrame()
    #dropping flights other than MCO, ORD, and JFK 
    # to keep additional filters (jetblue['Origin Airport'] == 'MCO') | (jetblue['Origin Airport'] == 'ORD') |
    filtered_data  = raw_df[(raw_df['Origin Airport'] == 'MCO') | (raw_df['Origin Airport'] == 'JFK') | (raw_df['Origin Airport'] == 'ORD')].copy()
    # filtered_data['Origin Airport'].unique()
    airline_df['DATE'] = filtered_data['Date (MM/DD/YYYY)']
    airline_df['DAY'] = [findDayOfWeek(date) for date in filtered_data['Date (MM/DD/YYYY)']]
    
    filtered_data['Flight Number'] = [flightNoStr(flightNo) for flightNo in filtered_data['Flight Number']]
    filtered_data['Flight No'] = filtered_data['Carrier Code'].str.cat(filtered_data['Flight Number'], sep=' ')
    airline_df['FLIGHT NUMBER'] = filtered_data['Flight No']

    airline_df['ORIGIN'] = filtered_data['Origin Airport']
    
    #converting time data to minutes
    filtered_data['Scheduled Arrival Time'] = [to_minutes(s) for s in filtered_data['Scheduled Arrival Time']] #to_minutes(step_1['Scheduled Arrival Time'])
    filtered_data['Actual Arrival Time'] = [to_minutes(s) for s in filtered_data['Actual Arrival Time']]
    filtered_data['Wheels-on Time'] = [to_minutes(s) for s in filtered_data['Wheels-on Time']]

    filtered_data['Scheduled Departure Time'] = filtered_data['Scheduled Arrival Time'] - filtered_data['Scheduled Elapsed Time (Minutes)']
    airline_df['DEPARTURE TIME'] = filtered_data['Scheduled Departure Time']
    airline_df['ARRIVAL TIME'] = filtered_data['Scheduled Arrival Time']
    # jetblue_df
    airline_df = airline_df[airline_df['ARRIVAL TIME']>=0] ##dropping data where flight departed one earlier date

    airline_df['CARRIER DELAY'] = filtered_data['Delay Carrier (Minutes)']
    airline_df['WEATHER DELAY'] = filtered_data['Delay Weather (Minutes)']

    airline_df['AIRPORT YEARLY ON-TIME DEP PERCENTAGE'] = [addDepPercentage(date, origin) for date, origin in zip(filtered_data['Date (MM/DD/YYYY)'], filtered_data['Origin Airport'])]
    
    airline_df['AIRLINE YEARLY ON-TIME ARR PERCENTAGE'] = [addArrPercentage(date, carrier_code) for date, carrier_code in zip(filtered_data['Date (MM/DD/YYYY)'], filtered_data['Carrier Code'])]

    # filtered_data['Arrival Delay (Minutes)']
    airline_df['ARRIVAL STATUS'] = [classifyDelay(delay) for delay in filtered_data['Arrival Delay (Minutes)']]

    # airline_df
    dfList.append(airline_df)
    
    # break
    # raw_data = pd.read_csv(file)
combined_df=pd.concat(dfList, ignore_index=True)
combined_df['DEPARTURE TIME'] = combined_df['DEPARTURE TIME'].astype(int)
# combined_df.to_csv('data/combined_arrival_data.csv', index=False)
combined_df
```
